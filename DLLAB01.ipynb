```python
import tensorflow as tf
import numpy as np
print(tf.__version__)
```

```python
#Create a scalar (0-D tensor)
a=tf.constant(5)
print(a)
```

```python
#Create a vector (1-D tensor)
v = tf.constant([1, 2, 3, 4])
print(v)
```

```python
# Create a matrix (2-D tensor)
m = tf.constant([[1, 2],
                 [3, 4]])
print(m)
```

```python
#Create a 3-D tensor
n=tf.constant([[1,2],[3,4],[5,6]])
print(n)
```

```python
#Specify data type explicitly
x = tf.constant([1, 2, 3], dtype=tf.float32)
print(x)
```

```python

z = tf.zeros((2, 3))
print(z)
```

```python
o = tf.ones((3, 2))
print(o)
```

```python
r = tf.range(1, 10, 2)
print(r)
```

```python
rn = tf.random.normal((2, 2))
print(rn)
```

```python
ru = tf.random.uniform((2, 2), minval=0, maxval=10)
print(ru)
```

```python
#Convert to NumPy
print(v.numpy())
```

```python
print("Shape:", m.shape)
print("Dtype:", m.dtype)
print("Rank:", tf.rank(m))
```

```python
a = tf.constant([10, 20, 30])
b = tf.constant([1, 2,3])
```

```python
# Addition of tensors
c = tf.add(a, b)
print(c)
```

```python
print(a+b)
```

```python
#Subtraction of tensors
d = tf.subtract(a, b)
print(d)
```

```python
print(a-b)
```

```python
#Element-wise multiplication
e=tf.multiply(a,b)
print(e)
```

```python
print(a * b)
```

```python
# Division of tensors
f = tf.divide(a, b)
print(f)
```

```python
print(a / b)
```

```python
# Floor division (integer division)
g = tf.math.floordiv(a, b)
print(g)
```

```python
#Power operation
p = tf.pow(b, 2)
print(p)
```

```python
#Matrix addition (same shape)
m1 = tf.constant([[1, 2],
                  [3,
4]])
m2 = tf.constant([[5, 6],
                  [7,
8]])
print(m1 + m2)
```

```python
t = tf.constant([[1, 2, 3],
                 [4,5,6]])
```

```python
print(t.dtype, t.shape, t.ndim)
```

```python
np_array = np.array([[1, 2, 3],
                     [4, 5, 6]])
```
